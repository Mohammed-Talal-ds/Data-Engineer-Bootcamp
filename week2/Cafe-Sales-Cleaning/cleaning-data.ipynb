### We need to convert the UNKNOWN and ERROR values to NaN to treat them the same

```python
df.replace(["UNKNOWN", "ERROR"], np.nan, inplace=True)
```

### Fill Price based on most frequent Price within same Item

```python
df["Price Per Unit"] = df["Price Per Unit"].fillna(
    df.groupby("Item")["Price Per Unit"]
      .transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
```

### Fill Items based on most frequent item within same Price

```python
df["Item"] = df["Item"].fillna(
    df.groupby("Price Per Unit")["Item"]
      .transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
print(df['Item'].isna().sum())
print(df['Price Per Unit'].isna().sum())
```

` We can Note that we still have NaNs in Item and Price  We will handle it later`

```python
df['Payment Method'].value_counts()
```

### Converting Types 

```python
df['Item'] = df['Item'].astype('category')
df['Quantity'] = df['Quantity'].astype('float32')
df['Price Per Unit'] = df['Price Per Unit'].astype('float32')
df['Total Spent'] = df['Total Spent'].astype('float32')
df['Location'] = df['Location'].astype('category')
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
df['Payment Method'] = df['Payment Method'].astype('category')
df.info()
```

### Fill the Total Spent with the multi between the Quantity and Price Per Unit

```python
df['Total Spent'] = df['Total Spent'].fillna(
    df['Quantity'] * df['Price Per Unit']
)
```

```python
df[df['Total Spent'].isna()].head()
```

` We can see that we still have NaN values in Total spent cause we still had some NaNs in the  Qunantity and Price we can handle that later` 

```python
df['Quantity'] = df['Quantity'].fillna(
    df['Total Spent'] / df['Price Per Unit']
)
```

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna(
    df['Total Spent'] / df['Quantity']
)
```

```python
df['Total Spent'] = df['Total Spent'].fillna(
    df['Quantity'] * df['Price Per Unit']
)
```

```python
df["Price Per Unit"] = df["Price Per Unit"].fillna(
    df.groupby("Item")["Price Per Unit"]
      .transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
df["Item"] = df["Item"].astype('category')
df["Price Per Unit"] = df["Price Per Unit"].astype("float32")
df["Item"] = df["Item"].fillna(
    df.groupby("Price Per Unit")["Item"]
      .transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
df.isna().sum()
```

### Try to fill the remaining NaNs using realtionship between columns

```python
mask = df['Total Spent'].isna() & \
       df['Quantity'].notna() & \
       df['Price Per Unit'].notna()

df.loc[mask, 'Total Spent'] = (
    df.loc[mask, 'Quantity'] * df.loc[mask, 'Price Per Unit']
)
```

```python
mask = df['Quantity'].isna() & \
       df['Total Spent'].notna() & \
       df['Price Per Unit'].notna()

df.loc[mask, 'Quantity'] = (
    df.loc[mask, 'Total Spent'] / df.loc[mask, 'Price Per Unit']
)
```

```python
mask = df['Price Per Unit'].isna() & \
       df['Total Spent'].notna() & \
       df['Quantity'].notna()

df.loc[mask, 'Price Per Unit'] = (
    df.loc[mask, 'Total Spent'] / df.loc[mask, 'Quantity']
)
```

```python
df.isna().sum()
```

`There was no diffrence we will try to find anyother relationship otherwise we will drop them `

```python
print(df['Location'].value_counts())
print(20 * '#')
print(df['Payment Method'].value_counts())
```

`We can Note that the distributions are almost blanaced So we can't use the mode method So we will just fill  them with UNKNOWN `

```python
df["Payment Method"] = df["Payment Method"].cat.add_categories("Unknown")
df["Payment Method"] = df["Payment Method"].fillna("Unknown")
df["Location"] = df["Location"].cat.add_categories("Unknown")
df["Location"] = df["Location"].fillna("Unknown")
```

`Fill the dates using the forward method `

```python
df["Transaction Date"] = df["Transaction Date"].ffill().bfill()
```

```python
df.info()
```

```python
df.isna().sum()
```

### Note we can drop all the NaN values from our Data Frame

```python
df = df.dropna( subset=['Item', 'Total Spent', 'Quantity', 'Price Per Unit'])
```
